### 1.3 Twitter API

```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = tweet_arch.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
import json

lines = []

with open('tweet_json.txt', 'r') as tweets:
    
    for line in tweets:
        line_dict = json.loads(line.strip())
        
        data = {
            "id": line_dict["id"],
            "retweet_count": line_dict["retweet_count"],
            "favorite_count": line_dict["favorite_count"]
        }
        
        lines.append(data)

tweet_data = pd.DataFrame(lines, columns = ["id", "retweet_count", "favorite_count"])
```

#### Missing Data

To address the missing data we will have to tidy up the dataset first, this would mean merging the three dataset together.

```python
clean_tweet_data.rename(columns={'id':'tweet_id'}, inplace=True)
```

#### Data Tidiness

1) Merge the three dataframes to form one.

**Define**

The clean_tweet_arch and clean_tweet_image are merged together and then id in the tweet_data is changed to tweet_id so as to merge it with the clean_tweet_arch. 

**Code**

```python
#Merge the clean twitter archive data to the clean twitter image prediction dataset together
clean_tweet_arch = pd.merge(clean_tweet_arch, clean_tweet_image, on ='tweet_id', how = 'left')
```

```python
#Merge the combined, clean twitter archive data together with the data from the twitter API
clean_tweet_arch = pd.merge(clean_tweet_arch, clean_tweet_data, on ='tweet_id', how = 'left')
```

## Conclusion

From our analyses and visualization, we can see that the pupper had the highest value count and the least favorite  and retweet count.
While floofer had the least value count and still had the favorite and retweet count.
We can also see that those who had high favorite count also had high retweet count, this signifies a direct relationship between these two.
We can also see an increase in retweet and favorite counts as the rating numerator increased from zero to fifteen and then there was low retweet and favorite counts for those greater than fifteen.
